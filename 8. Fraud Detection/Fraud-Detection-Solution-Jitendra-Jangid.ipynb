### Problem Statement & Approach

**Problem Statment:** Newly registered users on zomato platform based on the features available in the data. You would have to build a classification model to predict whether this user is fraudulent or not.
What we have observed is that many times people created fraud accounts to get new user promos or collude with restaurants to place fraud orders.

**Data Dictionary:**

- userid : user_id generated by our system
- device_id : unique identifier for the device from which account has been created.
- android_id : If the device is android then android id as per the device.
- event_timestamp : time of registration
- ip_address : ip address of the network used by the device.
- y : 0/1. 1 : Fraud User, 0 : Non Fraud User

The outline of this notebook is as follows:

1. **Exploratory Data Analysis (EDA)** - Analyzing the given data having registered users and other information (device, ip address, etc.). 

Performed Following Data Preparation & Analysis steps on Zomato users dataset:

- Checked the missing value counts for given variables
- Target Variable Distribution (Fraud User / Non-Fraud User)
- Number of unique values per variable (userid, device_id, android_id, ip_address)
- Distribution of Android vs Non-Android Users (Assumption: If device id is not generated -> Non-Android User)
- Variable Statistics
    - Distribution/Histogram of value counts for userid (What are the most common userids?)
    - Distribution/Histogram of value counts for android_id (What are the most common android_id?)
    - Distribution/Histogram of value counts for ip_address (What are the most common ip_address?)
    - Distribution/Histogram of value counts for device_id (What are the most common device_id?)
- Time wise trends
    - Hourly Registration Frequency
    - Hourly Conversion Ratio
    - IP Address Trends vs Fraud Rate
    
    
2. **Feature Engineering** - Created following additional features using event_timestamp, ip_address, device_id, android_id:


- Event Time Stamp:
    - Day
    - Hour
    - Minute
    - Seconds
- Android Available Flag (If Android values are not null : 1 otherwise 0)
- Group By Aggregation Based Features
    - Variance in day, for ip_address-userid-device_id
    - Variance in hour, for ip_address-userid-android_id
    - Variance in hour, for ip_address-day-device_id
    - Count of device_id, for ip_address-day-hour
    - Count of device_id, for ip_address-userid
    - Count of device_id, for ip_address-userid-android_id
    - Count of device_id, for ip_address-userid-day-hour
    - Mean hour, for ip_address-userid-device_id
    - Average registration on zomato app by distinct users
    - How popular is the userid for a device_id and vice versa?
    - Group by Features (Unique/Count based) using ip_address, device_id, android_id columns
- Time Till Next Event Registration
    - How long it takes for a given ip_address-userid-device_id before they perform the next event registration?
- Device Registrations Before & After
    - Has the user previously or subsequently registered the exact same device?


3. **Predictive Modeling Pipeline (Fraud/Non-Fraud Classifier)** - Leveraged pycaret module to build a machine learning based fraud detection classifier (LightGBM). Created following classifiers:


- Baseline Classifier (Using only given features: 'userid', 'device_id', 'android_id','ip_address') - 10 Fold Cross Validation Results
    - *AUC = 0.55*
    - *Precision = 0.15*
    - *Recall = 0.0001*
    - *F1 Score = 0.0003*
    
- Model After Feature Engineering - 10 Fold Cross Validation Results
    - *AUC = 0.85*
    - *Precision = 0.71*
    - *Recall = 0.31*
    - *F1 Score = 0.43*

- Model After Fixing Imbalance using SMOTE - 10 Fold Cross Validation Results
    - *AUC = 0.84*
    - *Precision = 0.55*
    - *Recall = 0.42*
    - *F1 Score = 0.48*

```python
### General Functions ###

def mkdir(*args):
    """
    To create directory for a given path
    """
    path = os.path.join(*args)
    try: 
        os.makedirs(path)
    except OSError:
        if not os.path.isdir(path):
            raise
    return path

def neat_print_sperator(name):
    """
    A function to create sperations between displays and prints
    """
    spacing_len = 95 - len(name)
    print('{} {} {}'.format('_'*math.floor((spacing_len/2)), name, '_'*math.ceil(spacing_len/2)))
    return
```

```python
### Dataframe related functions ###

def load_dataframe(file_path, file_type = "csv"):
    """
    A function to load datasets in csv/xlsx format
    """
    if file_type=="csv":
        df = pd.read_csv(file_path, index_col = None, encoding = "ISO-8859-1")
    elif file_type=="xlsx":
        df = pd.read_excel(file_path, index_col = None)
    elif file_type=="parquet":
        df = pd.read_parquet(file_path, engine='pyarrow')
    return df

def save_dataframe(data_frame, file_path, file_type = "csv"):
    if file_type=="csv":
        data_frame.to_csv(file_path, index = False)
    elif file_type=="xlsx":
        data_frame.to_xlsx(file_path, index = False)

def gradient_dataframe(df, caption):
    """
    Converting dataframe into Gradient DataFrame
    """
    
    agg_df = df.style.set_table_styles(
            [{'selector': 'tr:nth-of-type(odd)',
              'props': [('background', '#eee'),
                       ('text-align', 'left')]}, 
             {'selector': 'tr:nth-of-type(even)',
              'props': [('background', 'white'), ('text-align', 'left')]},
             {'selector': 'th',
              'props': [('background', '#808080'), 
                        ('color', 'white'),
                        ('font-family', 'verdana'),
                       ('text-align', 'left')]},
             {'selector': 'td',
              'props': [('font-family', 'verdana'), ('text-align', 'left')]},
            ]
            ).hide_index().format(formatter=None).set_caption(f'{caption}')\
                .background_gradient(cmap='Blues')
    return agg_df

def missing_value_of_data(data):
    """
    Function to find out missing values in the dataset
    """
    total=data.isnull().sum().sort_values(ascending=False)
    percentage=np.round(total/data.shape[0]*100,2)
    return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])
```

```python
### Plotly related functions ###

def plotly_bar_pie_chart(cnt_srs, layout_title, pie_title):
    """
    Plotly function to Plot Bar & Pie Charts
    """

    trace = go.Bar(
        x=cnt_srs.index,
        y=cnt_srs.values,
        marker=dict(
            color=cnt_srs.values,
            colorscale = 'Picnic',
            reversescale = True
        ),
    )

    layout = go.Layout(
        title=layout_title,
        font=dict(size=12)
    )

    data = [trace]
    fig = go.Figure(data=data, layout=layout)
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    py.iplot(fig, filename=pie_title)

    ## target distribution ##
    labels = (np.array(cnt_srs.index))
    sizes = (np.array((cnt_srs / cnt_srs.sum())*100))

    trace = go.Pie(labels=labels, values=sizes)
    layout = go.Layout(
        title=pie_title,
        font=dict(size=12)
    )
    data = [trace]
    fig = go.Figure(data=data, layout=layout)
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    py.iplot(fig, filename="usertype")

def horizontal_bar_chart(df, color):
    """
    Custom function for Horizontal Bar Chart
    """
    trace = go.Bar(
        y=df["word"].values[::-1],
        x=df["wordcount"].values[::-1],
        showlegend=False,
        orientation = 'h',
        marker=dict(
            color=color,
        ),
    )
    return trace

def plotly_box_plot(X, Y, caption, color):
    """
    Custom function for Box Plot
    """
    data = [go.Box(
            y=Y,
            x=X,
            showlegend=False,
            marker= dict(color=color)
        )]

    layout = go.Layout(
        title=caption
    )
    
    fig = go.Figure(data=data, layout=layout)
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    py.iplot(fig, filename='basic-box')

def plotly_histogram_chart(X, x_title, color):
    """
    Histogram chart of Continuous variable using Plotly
    """
    fig = go.Figure()
    fig.add_trace(go.Histogram(x=X, marker_color=color))
    # Overlay both histograms
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig.update_layout(
        title_text=f"{x_title} Distribution", # title of plot
        xaxis_title_text=x_title, # xaxis label
        yaxis_title_text='Count', # yaxis label
        title_x=0.5,
        barmode='overlay')

    # Reduce opacity to see both histograms
    fig.update_traces(opacity=0.75)
    fig.show()

def plotly_bubble_chart(X, Y, label):
    """
    Bubble chart using Plotly
    """
    fig = go.Figure(data=go.Scatter(
                    x=X,
                    y=Y,
                    hovertext = label,
                    mode='markers',
                    marker=dict(
                         color='rgb(255, 178, 102)',
                         size=10,
                         line=dict(
                            color='DarkSlateGrey',
                            width=1
                          )
                   )
    ))
    
    fig.update_xaxes(showline=True, linewidth=2, linecolor='black', mirror=True)
    fig.update_layout(
        title=f'Rating vs # Votes vs Cost',
        xaxis_title='Rating',
        yaxis_title='# Votes'
    )
    fig.show()
    
def plot_wordcloud(text, mask=None, max_words=200, max_font_size=100, figure_size=(24.0,16.0), 
                   title = None, title_size=40, image_color=False):
    """Function to Plot Word Clouds"""

    stopwords = set(STOPWORDS)
    more_stopwords = {'one', 'br', 'Po', 'th', 'sayi', 'fo', 'Unknown'}
    stopwords = stopwords.union(more_stopwords)

    wordcloud = WordCloud(background_color='black',
                    stopwords = stopwords,
                    max_words = max_words,
                    max_font_size = max_font_size, 
                    random_state = 42,
                    width=800, 
                    height=400,
                    mask = mask)
    wordcloud.generate(str(text))
    
    plt.figure(figsize=figure_size)
    if image_color:
        image_colors = ImageColorGenerator(mask);
        plt.imshow(wordcloud.recolor(color_func=image_colors), interpolation="bilinear");
        plt.title(title, fontdict={'size': title_size,  
                                  'verticalalignment': 'bottom'})
    else:
        plt.imshow(wordcloud);
        plt.title(title, fontdict={'size': title_size, 'color': 'black', 
                                  'verticalalignment': 'bottom'})
    plt.axis('off');
    plt.tight_layout()  
```

```python
### Text Pre-Processing Related Functions ###
def lemmatize_words(text):
    pos_tagged_text = nltk.pos_tag(text.split())
    return " ".join([lemmatizer.lemmatize(word, wordnet_map.get(pos[0], wordnet.NOUN)) for word, pos in pos_tagged_text])

def generate_ngrams(text, n_gram=1):
    """
    Custom function for ngram generation
    """
    token = [token for token in str(text).lower().split(" ") if token != "" if token not in STOPWORDS]
    ngrams = zip(*[token[i:] for i in range(n_gram)])
    return [" ".join(ngram) for ngram in ngrams]

def gensim_clean_text(s):
    """
    Clean the text using Gensim library
    """
    filters = [
               gsp.strip_tags, 
               gsp.strip_punctuation,
               gsp.strip_multiple_whitespaces,
               gsp.strip_numeric,
               gsp.remove_stopwords, 
               gsp.strip_short
              ]
    s = str(s).lower()
    s = utils.to_unicode(s)
    s = re.sub(r'[^\x00-\x7F]+',' ', s)
    s = re.sub('https?://\S+|www\.\S+', '', s)
    s = re.sub('\n', '', s)
    s = lemmatize_words(s)
    for f in filters:
        s = f(s)
    return s
```

#### Group By - Aggregation Based Features

```python
"""
Let us first fill the null values which are available in device_id, android_id, ip_address
"""
user_df["device_id"] = user_df["device_id"].fillna("123456789")
user_df["android_id"] = user_df["android_id"].fillna("987654321")
user_df["ip_address"] = user_df["ip_address"].fillna("123456789987654321")
```

```python
"""
First Let us Define all the groupby aggregation transformations
"""
GROUPBY_AGGREGATIONS = [
    
    # V1 - Group by Features #
    ##########################   
    # Variance in day, for ip_address-userid-device_id
    {'groupby': ['ip_address','userid','device_id'], 'select': 'day', 'agg': 'var'},
    # Variance in hour, for ip_address-userid-android_id
    {'groupby': ['ip_address','userid','android_id'], 'select': 'hour', 'agg': 'var'},
    # Variance in hour, for ip_address-day-device_id
    {'groupby': ['ip_address','day','device_id'], 'select': 'hour', 'agg': 'var'},
    # Count of device_id, for ip_address-day-hour
    {'groupby': ['ip_address','day','hour'], 'select': 'device_id', 'agg': 'count'},
    # Count of device_id, for ip_address-userid
    {'groupby': ['ip_address', 'userid'], 'select': 'device_id', 'agg': 'count'},        
    # Count of device_id, for ip_address-userid-android_id
    {'groupby': ['ip_address', 'userid', 'android_id'], 'select': 'device_id', 'agg': 'count'},
    # Count of device_id, for ip_address-userid-day-hour
    {'groupby': ['ip_address','userid','day','hour'], 'select': 'device_id', 'agg': 'count'},
    # Mean hour, for ip_address-userid-device_id
    {'groupby': ['ip_address','userid','device_id'], 'select': 'hour', 'agg': 'mean'}, 
    
    # V2 - Group by Features #
    ##########################
    # Average registration on zomato app by distinct users
    {'groupby': ['userid'], 
     'select': 'ip_address', 
     'agg': lambda x: float(len(x)) / len(x.unique()), 
     'agg_name': 'AvgViewPerDistinct'
    },
    # How popular is the userid for a device_id?
    {'groupby': ['userid'], 'select': 'device_id', 'agg': 'count'},
    {'groupby': ['device_id'], 'select': 'userid', 'agg': 'count'},
    
    # V3 - Group by Features (Unique/Count based) #                                  
    ############################################### 
    {'groupby': ['ip_address'], 'select': 'device_id', 'agg': 'nunique'}, 
    {'groupby': ['ip_address'], 'select': 'userid', 'agg': 'nunique'}, 
    {'groupby': ['ip_address','day'], 'select': 'hour', 'agg': 'nunique'}, 
    {'groupby': ['ip_address','userid'], 'select': 'android_id', 'agg': 'nunique'}, 
    {'groupby': ['userid'], 'select': 'device_id', 'agg': 'nunique'}, 
    {'groupby': ['ip_address', 'device_id', 'android_id'], 'select': 'userid', 'agg': 'nunique'}, 
    {'groupby': ['ip_address','device_id','android_id'], 'select': 'userid', 'agg': 'cumcount'}, 
    {'groupby': ['ip_address'], 'select': 'userid', 'agg': 'cumcount'}, 
    {'groupby': ['ip_address'], 'select': 'android_id', 'agg': 'cumcount'}, 
]

# Apply all the groupby transformations
for spec in GROUPBY_AGGREGATIONS:
    
    # Name of the aggregation we're applying
    agg_name = spec['agg_name'] if 'agg_name' in spec else spec['agg']
    
    # Name of new feature
    new_feature = '{}_{}_{}'.format('_'.join(spec['groupby']), agg_name, spec['select'])
    
    # Print statments
    print(">> Grouping by {}, and aggregating {} with {}".format(
        spec['groupby'], spec['select'], agg_name
    ))
    
    # Unique list of features to select
    all_features = list(set(spec['groupby'] + [spec['select']]))
    
    # Perform the groupby
    gp = user_df[all_features]. \
        groupby(spec['groupby'])[spec['select']]. \
        agg(spec['agg']). \
        reset_index(). \
        rename(index=str, columns={spec['select']: new_feature})
        
    # Merge back to original data
    if 'cumcount' == spec['agg']:
        user_df[new_feature] = gp[0].values
    else:
        user_df = user_df.merge(gp, on=spec['groupby'], how='left')
        
    # Clear memory
    del gp
    gc.collect()
```

#### Time Till Next Event Registration

How long it takes for a given ip_address-userid-device_id before they perform the next event registration?

```python
GROUP_BY_NEXT_EVENTS = [
    
    # V1
    {'groupby': ['ip_address']},
    {'groupby': ['ip_address', 'userid']},
    {'groupby': ['ip_address', 'device_id']},
    {'groupby': ['ip_address', 'android_id']},
    
    # V2
    {'groupby': ['ip_address', 'userid', 'android_id', 'device_id']},
    {'groupby': ['ip_address', 'android_id', 'device_id']},
]

# Calculate the time to next event for each group
for spec in GROUP_BY_NEXT_EVENTS:
    
    # Name of new feature
    new_feature = '{}_nextEvent'.format('_'.join(spec['groupby']))    
    
    # Unique list of features to select
    all_features = spec['groupby'] + ['event_timestamp']
    
    # Run calculation
    print(f">> Grouping by {spec['groupby']}, and saving time to next event in: {new_feature}")
    user_df[new_feature] = user_df[all_features].groupby(spec['groupby']).event_timestamp.transform(lambda x: x.diff().shift(-1)).dt.seconds
```

#### Device Registrations Before & After 

Has the user previously or subsequently registered the exact same device?

```python
HISTORY_EVENTS = {
    'identical_events': ['ip_address', 'userid', 'device_id', 'android_id'],
    'app_events': ['ip_address', 'userid']
}

# Go through different group-by combinations
for fname, fset in HISTORY_EVENTS.items():
    
    # Registration in the past
    user_df['prev_'+fname] = user_df. \
        groupby(fset). \
        cumcount(). \
        rename('prev_'+fname)
        
    # Registration in the future
    user_df['future_'+fname] = user_df.iloc[::-1]. \
        groupby(fset). \
        cumcount(). \
        rename('future_'+fname).iloc[::-1]
```
