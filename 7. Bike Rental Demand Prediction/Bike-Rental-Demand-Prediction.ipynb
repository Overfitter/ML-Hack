```python
# Generate a function to create more distinctive print output
def neat_print_sperator(name):
    '''A function to create sperations between displays and prints'''
    spacing_len = 95 - len(name)
    print('{} {} {}'.format('_'*math.floor((spacing_len/2)), name, '_'*math.ceil(spacing_len/2)))
    return
```

### Function & Utilities

```python
def missing_value_of_data(data):
    """
    Function to find out missing values in the dataset
    """
    total=data.isnull().sum().sort_values(ascending=False)
    percentage=np.round(total/data.shape[0]*100,2)
    return pd.concat([total,percentage],axis=1,keys=['Total','Percentage'])
```

```python
## Function for calculation of evaluation metric
def mean_absolute_percentage_error(y_true, y_pred):
    mask = y_true != 0
    return (np.fabs(y_true - y_pred)/y_true)[mask].mean()

def median_absolute_percentage_error(y_true, y_pred):
    mask = y_true != 0
    return (np.fabs(y_true - y_pred)/y_true)[mask].median()

def mean_absolute_error_overall(y_true, y_pred):
    return np.fabs(y_true - y_pred).mean()

def mean_absolute_error_zero(y_true, y_pred):
    try:
        mask = y_true == 0
        return np.fabs(y_true - y_pred)[mask].mean()
    except:
        return "error"

def mean_absolute_error_nonzero(y_true, y_pred):
    mask = y_true != 0
    return np.fabs(y_true - y_pred)[mask].mean()

def rmse_overall(y_true, y_pred):
    #mask = y_true != 0
    temp_y = y_true
    temp_p = y_pred
    return mean_squared_error(temp_y, temp_p)**0.5

def rmse_zero(y_true, y_pred):
    try:
        mask = y_true == 0
        temp_y = y_true[mask]
        temp_p = y_pred[mask]
        return mean_squared_error(temp_y, temp_p)**0.5
    except:
        return 'error'

def rmse_nonzero(y_true, y_pred):
    mask = y_true != 0
    temp_y = y_true[mask]
    temp_p = y_pred[mask]
    return mean_squared_error(temp_y, temp_p)**0.5

def median_absolute_error_mp(y_true, y_pred):
    return np.fabs(y_true - y_pred).median()

def avg_absolute_predicted_for_zero(y_true, y_pred):
    mask = y_true == 0
    return np.fabs(y_pred)[mask].mean()

def max_predicted_for_zero(y_true, y_pred):
    try:
        mask = y_true == 0
        return y_pred[mask].max()
    except:
        return 'error'

def min_predicted_for_zero(y_true, y_pred):
    try:
        mask = y_true == 0
        return y_pred[mask].min()
    except:
        return 'error'

def evaluation_metrics(y_true, y_pred):
    """
    """
    temp = pd.DataFrame({'y_test':y_true, 'pred':y_pred})
    temp = temp.loc[temp['y_test']>0]
    temp['diff'] = temp['y_test'] - temp['pred']
    temp['perc'] = temp['diff']/temp['y_test']    
    
    print("MAE")
    print("\tMAE overall : " + str(mean_absolute_error_overall(y_true, y_pred)))
    print("\tMAE nonzero : " + str(mean_absolute_error_nonzero(y_true, y_pred)))
    print("\tMAE zero    : " + str(mean_absolute_error_zero(y_true, y_pred)))
    print("\nMAPE")
    print("\tMAPE nonzero: " + str(mean_absolute_percentage_error(y_true, y_pred)))
    print("\nRMSE")
    print("\tRMSE overall: " + str(rmse_overall(y_true, y_pred)))
    print("\tRMSE nonzero: " + str(rmse_nonzero(y_true, y_pred)))
    print("\tRMSE zero   : " + str(rmse_zero(y_true, y_pred)))
    print("R-Squared : " +  str(r2_score(y_true, y_pred)))
```

```python
def run_pipeline(sklearn_model, train, test, target_var, exclude_other_vars=[]):
    """
    Function to Run a Scikit-Learn Models and Get the train & test predictions with Error Metrics
    """
    neat_print_sperator(f"Running Predictive Model {target_var}")
    print("\n")
    
    model_feature_cols = list(set(train.columns.tolist()) - set(exclude_other_vars))

    model_data = train[model_feature_cols + [target_var]]
    
    
    X_train = train.drop([target_var], axis = 1)
    y_train = train[target_var]
    
    X_test = test.drop([target_var], axis = 1)
    y_test = test[target_var]    
    
    # Add silent=True to avoid printing out updates with each cycle
    sklearn_model.fit(X_train, y_train)

    
    # make predictions (train)
    neat_print_sperator("Metrics on Train Dataset")
    predictions_train = sklearn_model.predict(X_train)
    predictions_train[predictions_train<0] = 0
    evaluation_metrics(y_train, predictions_train)

    neat_print_sperator("Metrics on Test Dataset")
    
    # make predictions (test)
    predictions_test = sklearn_model.predict(X_test)
    predictions_test[predictions_test<0] = 0
    evaluation_metrics(y_test, predictions_test)
    
    neat_print_sperator("Actual vs Prediction Plot")
    plt.figure(figsize=(12, 8))
    p = sn.scatterplot(x = predictions_test, 
                        y = y_test)
    p.set_ylabel('Actual', fontweight='bold')
    p.set_xlabel('Prediction', fontweight='bold')
  
    return sklearn_model, predictions_train, predictions_test
```

#### Creating Complete Data using Station, Trip & Weather Datasets

```python
## Let's first merge trip_data & station_data using Id & Start Station , End Station columns
trip_station_data = pd.merge(trip_data, station_data.add_prefix('Start Station '), left_on=["Start Station"], 
                           right_on=["Start Station Id"], how = "left").drop(["Start Station Id"], axis = 1)

trip_station_data = pd.merge(trip_station_data, station_data.add_prefix('End Station '), left_on=["End Station"], 
                           right_on=["End Station Id"], how = "left").drop(["End Station Id"], axis = 1)
```

```python
## First Clean the date columns of Trip & Station Data
trip_station_data["Start Date Only"] = trip_station_data["Start Date"].apply(lambda x: x.split(" ")[0])
trip_station_data["End Date Only"] = trip_station_data["End Date"].apply(lambda x: x.split(" ")[0])
```

```python
## In weather data, we have observed that for a given day we have multiple rows: roll up data on Date column & take average
## Ignoring Events & Zip columns because they are not relevant
weather_agg_df = weather_data.drop(["Events", "Zip"], axis=1).groupby(["Date"], as_index=False).agg(lambda x: x.mean(skipna=True))
```

```python
## Let's merge Weather Data with trip & station data to get weather information based on Start Date
complete_data = pd.merge(trip_station_data, weather_agg_df, left_on=["Start Date Only"], right_on = ["Date"], 
                       how = "left").drop(["Date"], axis=1)
```

```python
complete_data.head(2)
```

```python
## Now lets create the target variable : Net change in the stocks between two stations
complete_data["Net Dock Change"] = (complete_data['Start Station Dock Count'] - complete_data['End Station Dock Count']).abs()
```

### Feature Engineering

- Create new columns "date,"hour","weekDay","month" from "Start Date" column.
- Drop the datetime column as we already extracted useful features from it.
- Create Holiday & Working day variable

#### Creating New Columns From Date Column

```python
complete_data["Hour"] = complete_data["Start Date"].apply(lambda x : x.split()[1].split(":")[0])
complete_data["Weekday"] = complete_data["Start Date Only"].apply(lambda dateString : calendar.day_name[datetime.strptime(dateString,"%d/%m/%Y").weekday()])
complete_data["Month"] = complete_data["Start Date Only"].apply(lambda dateString : calendar.month_name[datetime.strptime(dateString,"%d/%m/%Y").month])
complete_data["Working Day"] = np.where(complete_data["Weekday"].isin(["Saturday", "Sunday"]),0,1)
```

```python
## Create Holiday feature
min_date = str(pd.to_datetime(complete_data["Start Date Only"]).min()).split()[0]
max_date = str(pd.to_datetime(complete_data["Start Date Only"]).max()).split()[0]

cal = Calendar()
holidays = cal.holidays(start=min_date, end=max_date)
holidays = [str(x).split()[0] for x in holidays]
holidays = [datetime.strptime(x, '%Y-%m-%d').strftime('%d/%m/%Y') for x in holidays]

complete_data['Holiday'] = np.where(complete_data["Start Date Only"].isin(holidays),1,0)
```

#### Converting into Category Type

```python
categoryVariableList = ["Hour","Weekday","Month","Holiday","Working Day","Subscriber Type"]
for var in categoryVariableList:
    complete_data[var] = complete_data[var].astype("category")
```
